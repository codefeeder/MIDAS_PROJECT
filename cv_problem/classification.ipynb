# Imports

```python
from fastai.vision import *
from fastai.metrics import error_rate
import pickle
import numpy as np
from pathlib import Path
```

```python
path = Path('../../../../data/midas data')
```

```python
path.ls()
```

# Loading the provided Pickle Data

```python
pickle_in = open(path/"train_image.pkl","rb")
train_image = pickle.load(pickle_in)
pickle_in = open(path/"test_image.pkl","rb")
test_image = pickle.load(pickle_in)
pickle_in = open(path/"train_label.pkl","rb")
train_label = pickle.load(pickle_in)
```

```python
classes = {}
for i in range(0,len(train_label)):
    if train_label[i] not in classes.keys():
        classes[train_label[i]] = i

print(classes)
```

#### The above analysis confirms that the problem is a classical classification problem with 4 classes and can be solved using a pretrained CNN

# Preparing Data for Creating Data Bunch

### Data Bunch can be created using data and labels in a variety of ways few of which are :-
#### 1) Labels provided as csv
#### 2) Folders named as the corresponding class name
#### 3) Labels provided as DataFrame

#### Here I choose to follow the most convinient method where images are saved in a folder named after their corresponding class labels

```python
from scipy.misc import imsave
```

```python
for i in range(0,len(train_image)):
    img = (np.reshape(train_image[i],(28,28)))
    plt.imsave(path/f'train/{train_label[i]}/{i}.jpg',img)
```

# Preparing DataBunch object 
### The DataBunch is loaded into the CNN model during initialization and contains the training and test data loaders 
### Parameters passed are path of training data and test data, resizing value and augmentations to applied
#### Augmentations are applied to create multiple images from a single image which helps the model to generalize better and also increases the size of training data. get_transforms is a unction that returns the augmented training data on the basis of parameters passed to the function

```python
data = ImageDataBunch.from_folder(path, train="path/train",test = "test", valid_pct=0.2,
        ds_tfms=get_transforms(max_zoom=0.5, max_warp=0.05, max_rotate=0.05, max_lighting=0.2), size=229, num_workers=4).normalize(imagenet_stats)
```

```python
data.classes
```

### An overview of the data that is contained in the DataBunch and will be used for training

```python
data.show_batch(rows=3, figsize=(7,8))
```

# Importing the pretrained resnet50 model

```python
learn = create_cnn(data, models.resnet50, metrics=error_rate)
```

```python
import gc
gc.collect()
```

# Training the model

```python
learn.fit_one_cycle(4)
```

```python
learn.save('model1')
```

```python
learn = learn.load('model1')
```

# Trying to find the best learning rate for the model
### A learning rate is termed as optimum if the learning is not too high so as to miss the global optima value and at the same time is not too slow such that it consumes large time to reach the global optimum.

### lr_find() trains the model at different learning rates and plots the learning rates against loss values
### The maximum rate at which the loss is still decreasing can be used for training the model

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10,1e-04)
```

```python
learn.save('model-2')
```

```python
learn = learn.load('model-2')
```

### Training without unfreezing updates activations for only last layer(head) of the CNN
### The model is pretrained on the imagenet dataset and the activations for previous layers works precisely well for all datasets but unfreezing and training helps to adjust the activations of even the previous layers to a more suitable value in accordance to our training dataset

```python
learn.unfreeze() 
learn.lr_find()
learn.recorder.plot()
```

### The previous layers are more or less perfect in their job such as detecting lines, detecting shapes and patterns and therefore, the earlier layers should be trained at a rate slower than the rate used for the later layers which are highly inefficient in their job
### We Progressively reduce the learning rate as we move towards the initial layers

```python
lr = 2e-05
lrs = [lr/100,lr/10,lr]
```

```python
learn.fit_one_cycle(20,lrs)
```

```python
learn.save('model-unfreeze-1')
```

```python
learn.freeze()
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(20,2e-04)
```

### As clearly visible that during the later epochs the model started to overfit(Training loss was decreasing while validation loss was constantly increasing ) therefore, we should try using different augmentations and bigger image size for further training

```python
data = ImageDataBunch.from_folder(path, train="path/train",test = "test", valid_pct=0.2,
        ds_tfms=get_transforms(max_zoom=0.5, max_warp=0.05, max_rotate=0.05, max_lighting=0.2),bs=24, size=299, num_workers=4).normalize(imagenet_stats)
learn.data = data
```

```python
learn.freeze() #Freezes all the layers except the head and therefore, activations for only head can be updated
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(20,1e-04)
```

```python
learn.save('model-1-299')
```

```python
learn.unfreeze()
learn.lr_find()
learn.recorder.plot()
```

```python
lr = 6e-07
lrs = [lr/100,lr/10,lr]
```

```python
learn.fit_one_cycle(10,lrs)
```

```python
learn.save('model-299-unfreezed-1')
```

```python
learn.freeze()
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10,5e-06)
```

```python
learn.save('model-fuf-299')
```

```python
learn = learn.load('model-fuf-299')
```

### Summary provides an overview of the total number of layers in our network and the shape and type of each layer 

```python
learn.summary()
```

```python
for i in range(0,len(test_image)):
    img = (np.reshape(test_image[i],(28,28)))
    plt.imsave(path/f'../test/{i}.jpg',img)
```

# Analysing the performance of the model on the randomly generated validation set

```python
result_interpretation = ClassificationInterpretation.from_learner(learn)
```

```python
result_interpretation.plot_confusion_matrix()
```

### According to the previous tables the validation loss was less than the training loss that signifies underfitting but even several epochs and training freezed model didn't help much and therefore, it emphasizes on the need of running few more epochs on unfreezed model to find the best fitting set of activation energies

```python
learn.unfreeze()
learn.lr_find()
learn.recorder.plot()
```

```python
lr = 5e-06
lrs = [lr/100,lr/10,lr]
```

```python
learn.fit_one_cycle(4,lrs)
```

```python
learn.save('final-model')
```

```python
result_interpretation = ClassificationInterpretation.from_learner(learn)
result_interpretation.plot_confusion_matrix()
```

### Due to the approach followed above the number of misclassified instances reduced steeply (Diagonal represents the number of correctly classified instances while all others are misclassified)
## The final model has an error_rate of 0.029

```python
learn = learn.load('final-model')
```
